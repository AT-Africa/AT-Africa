### CALCULATING HML AND SMB
#### BOOK-TO-MARKET AND MARKET-CAP

     Market Capitalization= Price * Total shares outstanding

     Price-to-book = Market cap/ (total assets - total liabilities)

     book-to-market-ratio=1/Price-to-book

     book-to-market-ratio = (total assets - total liabilities) / Market cap

```python
def calculate_values(filepath):
    """A function that imports the datasets,resamples from monthly to weekly and calculates it's book-to-market ratio and market capitalization"""
    df=pd.read_excel(filepath)
    
    df.rename (columns= {df.columns[1]:'prices',
                         df.columns [2]: 'OS',
                         df.columns[3]:'TA',
                         df.columns[4]:'TL'},inplace=True)

    df['Date']=pd.to_datetime(df['Date'])

    df['n1']=1

    df=df.set_index('Date')

    df.index=df.index.map(lambda d: d - pd.tseries.offsets.Day(d.weekday()))

    dti=pd.date_range(df.index.min(),df.index.max(),freq='7D',name='Date')

    df=df.pivot(columns='n1',values=['prices','OS','TA','TL']).reindex(dti)

    df=df.fillna(method='ffill')

    df.columns=['prices','OS','TA','TL']

    df['market_cap']=(df.iloc[:,0]) * (df.iloc[:,1])
    df['book_to_market_ratio']=((df.iloc[:,2]) - (df.iloc[:,3])) / df['market_cap']
    
    df=df.shift(1)
    
    df=df.iloc[1:,:]
    return df
```

```python
book_to_market_values=[]
for df in list1:
    bmr = df.sort_values(by='Book_to_market_ratio',ascending=True)
    book_to_market_values.append(bmr)
```

```python
market_Cap=[]
for i in list1:
    Market_Cap = i.sort_values(by='Market_Cap',ascending=True)
    market_Cap.append(Market_Cap)
```

```python
final_Returns=[]
for l in list1:
    my_returns = l['Change_in_Price']
    final_Returns.append(my_returns)
```

```python
sc = []
bc = []
for idx,i in enumerate(market_Cap):
    half=np.percentile(i['Market_Cap'].dropna(),50)#50th percentile of market cap\n",
    small_caps=list1[idx][i['Market_Cap']<half]#market caps below the 50th percentile\n",
    big_caps=list1[idx][i['Market_Cap']>half]#market caps above the 50th percentile\n",
    sc.append(small_caps)
    bc.append(big_caps)
```

```python
val=[]
neu=[]
gr=[]
for idx,i in enumerate(book_to_market_values):
    thirty=np.percentile(i['Book_to_market_ratio'].dropna(),30)
    seventy=np.percentile(i['Book_to_market_ratio'].dropna(),70)

    growth=list1[idx][i['Book_to_market_ratio']<=thirty]
    neutral=list1[idx][(i['Book_to_market_ratio']>=thirty) & (i['Book_to_market_ratio']<=seventy)]
    value=list1[idx][i['Book_to_market_ratio']>=seventy]
    val.append(value)
    neu.append(neutral)
    gr.append(growth)
```

```python
hml=[]
smb=[]
for i in range(47):
    small_value=sc[i].index.intersection(val[i].index)
    small_neutral=sc[i].index.intersection(neu[i].index)
    small_growth=sc[i].index.intersection(gr[i].index)
    
    big_value=bc[i].index.intersection(val[i].index)
    big_neutral=bc[i].index.intersection(neu[i].index)
    big_growth=bc[i].index.intersection(gr[i].index)
    
    sv=final_Returns[i][small_value].mean()
    sn=final_Returns[i][small_neutral].mean()
    sg=final_Returns[i][small_growth].mean()
    
    bv=final_Returns[i][big_value].mean()
    bn=final_Returns[i][big_neutral].mean()
    bg=final_Returns[i][big_growth].mean()
    HML=(1/3*(sv+sn+sg)) - (1/3*(bv + bn + bg))
    SMB=(1/2*(sv + bv)) - (1/2*(sg +bg ))
    hml.append(HML)
    smb.append(SMB)
    #print(sv,sn,sg,bv,bn,bg)
    #print(HML,SMB)
```
